```python

import pandas as pd
from config import gkey
import requests
from pprint import pprint
```

```python
zillow = pd.read_csv("Zillow_Metro.csv")
```

```python
zillow["Name"] = ""
zillow["lat"] = ""
zillow["lng"] = ""

params = {"key": gkey}

counter = 0
for index, row in zillow.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json?"

    this = row["RegionName"]

    #params['componenetRestrictions'] = f'administrative_area_1: "{this}"'
    #params['components'] = f'administrative_area_level_1: OH'
    params["address"] = this
    query = requests.get(base_url, params=params)
    
    data = query.json()
    
    try:
        lat = data["results"][0]["geometry"]["location"]["lat"]
        lng = data["results"][0]["geometry"]["location"]["lng"]
        
        zillow.loc[index, "lat"] = lat
        zillow.loc[index, "lng"] = lng
    except:
        print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!Character not found")
        pass
    counter = counter + 1
    #print(long_county[0], counter)
    print(lat, lng, counter)
    
zillow
    
    
zillow.to_csv("zillow_w_lat_lng")
```
