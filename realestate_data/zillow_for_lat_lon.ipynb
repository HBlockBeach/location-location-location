```python
import pandas as pd
from config import gkey
import requests
from pprint import pprint
```

```python
zillow = pd.read_csv("Zillow_Metro.csv")

zillow
```

```python

zillow["lat"] = ""
zillow["lng"] = ""

params = {"key": gkey}

counter = 0
for index, row in zillow.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json?"

    this = row["RegionName"]
    params["address"] = this
    
    query = requests.get(base_url, params=params)
    data = query.json()
    
    try:
        lat = data["results"][0]["geometry"]["location"]["lat"]
        lng = data["results"][0]["geometry"]["location"]["lng"]
        
        zillow.loc[index, "lat"] = lat
        zillow.loc[index, "lng"] = lng

    except:
        print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!Character not found")
        pass
    
    counter = counter + 1
  
    print(lat, lng, counter)
    
```

```python
zillow= zillow[["RegionName", "StateName", "1/31/2019", "2/28/2019", "3/31/2019","4/30/2019","5/31/2019","6/30/2019", "7/31/2019","8/31/2019", "9/30/2019", "10/31/2019", "11/30/2019", "12/31/2019", "1/31/2020", "2/29/2020", "3/31/2020", "4/30/2020", "5/31/2020", "lat", "lng"]]
 
zillow = zillow.drop(zillow.index[0])
 
zillow.to_csv("zillow_w_lat_lng.csv", sep="\t")
```
